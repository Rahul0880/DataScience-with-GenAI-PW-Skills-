# 1. What is a parameter?
"""
A parameter is a variable that defines a model’s behavior during training.
In Machine Learning, parameters are learned from data (e.g., weights and biases in neural networks or coefficients in linear regression).
"""

# 2. What is correlation?
#    What does negative correlation mean?
"""
Correlation measures the strength and direction of a linear relationship between two variables.
It ranges from -1 to +1.
- Positive correlation: both variables move in the same direction.
- Negative correlation: one increases while the other decreases.
"""

# 3. Define Machine Learning. What are the main components in Machine Learning?
"""
Machine Learning is a field of Artificial Intelligence that enables systems to learn from data and make predictions without explicit programming.

Main components:
1. Dataset (input data)
2. Model (algorithm)
3. Objective or Loss function
4. Optimization algorithm (training process)
"""

# 4. How does loss value help in determining whether the model is good or not?
"""
Loss indicates how far the model’s predictions deviate from actual values.
A smaller loss means better performance.
High loss = poor predictions, low loss = good predictions.
"""

# 5. What are continuous and categorical variables?
"""
- Continuous variables: numeric values that can take infinite possible values within a range (e.g., temperature, height, age).
- Categorical variables: qualitative data divided into discrete categories (e.g., gender, color, type).
"""

# 6. How do we handle categorical variables in Machine Learning? What are the common techniques?
"""
Categorical variables are converted into numerical format for model training using:
1. Label Encoding – assigns integer labels to categories.
2. One-Hot Encoding – creates binary columns for each category.
3. Target Encoding – replaces each category with the mean of target values.
"""

# 7. What do you mean by training and testing a dataset?
"""
Training dataset – used to train and fit the model.
Testing dataset – used to evaluate the trained model’s performance on unseen data.
This ensures that the model generalizes well.
"""

# 15. What is causation? Explain difference between correlation and causation with an example.
"""
Causation means one variable directly influences or causes a change in another.
Correlation only means the two variables are related, not that one causes the other.

Example:
Ice cream sales and drowning incidents are correlated, but increased temperature causes both (not each other).
"""

# 16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
"""
An optimizer updates model parameters to minimize loss during training.

Types:
1. Gradient Descent – updates weights using full dataset.
2. Stochastic Gradient Descent (SGD) – updates weights per sample.
3. Adam – adaptive learning rate optimization (commonly used).
4. RMSProp, Adagrad – variants that improve convergence.

Example:
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
"""
